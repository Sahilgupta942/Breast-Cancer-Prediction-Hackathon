```python
scaler=MinMaxScaler()
final=pd.DataFrame(scaler.fit_transform(df.drop('diagnosis_M',axis=1)),columns=df.columns[1:])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(final,df['diagnosis_M'],test_size=0.3,random_state=101)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
cross_val_score(DecisionTreeClassifier(),final,df['diagnosis_M'],cv=3)
```

```python
cross_val_score(LogisticRegression(),final,df['diagnosis_M'],cv=3)
```

```python
cross_val_score(SVC(),final,df['diagnosis_M'],cv=3)
```

```python
model_params={
    'SVM':{
        'model':SVC(gamma='auto'),
        'params':{
        'C':[0.1,1,10,20],
        'kernel':['rbf','linear']
    }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
        'n_estimators':[10,30,50]
    }
    },
    'Logistic_Regression':{
    'model':LogisticRegression(),
    'params':{
    'C':[1,5,10]
    }
    }
}
```

```python
score=[]
for model_name,model_pm in model_params.items():
    grid=GridSearchCV(model_pm.get('model'),model_pm.get('params'))
    grid.fit(x_train,y_train)
    score.append({
        'model':model_name,
        'model_score':grid.best_score_,
        'best_params':grid.best_params_
    })
```

```python
score
```

```python
final=SVC(C=20,kernel='rbf')
final.fit(x_train,y_train)
```

```python
pred=final.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,pred))
```

```python
print(metrics.confusion_matrix(y_test,pred))
```
